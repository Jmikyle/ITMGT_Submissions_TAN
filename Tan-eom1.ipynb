```python
#Turning Tables
rows = int(input("Enter number of rows:"))
print("<table>")
for row in range(0,rows):
    row = row+1
    if row%2 == 0:
        color = '"whiterow"'
    else:
        color = '"lightyellowrow"'
    square = str(row**2)
    line = f"<tr class{color}><td>{row}</td><td>{square}</td></tr>"
    print(line)
print("</table>")
```
